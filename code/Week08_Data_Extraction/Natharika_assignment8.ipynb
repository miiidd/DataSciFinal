```python
import dateparser
```

To convert from thai text date string, we will use the parse method.  Note that the parse method assumes the BC year, not BE.  Thus, we will have to subtract 543 from the year.  In addition, weekday() returns day of week with 0=Monday, ..., 6=Sunday.

```python
dt = dateparser.parse('วันศุกร์ที่ 17 มกราคม 2563')

# this will print out weekday == 0 (Monday)
print(dt)
print(dt.weekday())

# this will print out weekday == 4 (Friday)
dt = dt.replace(year=dt.year-543)
print(dt)
print(dt.weekday())
```

```python
dt = dateparser.parse('วันเสาร์ที่ 21 กันยายน 2564')
dt = dt.replace(year=dt.year-543)
print(dt)
print(dt.weekday())
```

Count the distribution of number of week days that are “วันพระ” for all three years and answer the following questions:

```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
buddhist1 = requests.get('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2565.aspx')
buddhist2 = requests.get('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2566.aspx')
buddhist3 = requests.get('https://www.myhora.com/ปฏิทิน/วันพระ-พ.ศ.2567.aspx')
```

```python
soup1 = BeautifulSoup(buddhist1.text, "lxml")
soup2 = BeautifulSoup(buddhist2.text, "lxml")
soup3 = BeautifulSoup(buddhist3.text, "lxml")
```

## How many วันพระ in total (of 3 years)?

```python
bud_list = soup1.select('div.bud-day')+soup2.select('div.bud-day')+soup3.select('div.bud-day')
len(bud_list)
```

**ANSWER** 152 days

## How many days in total (of 3 years) that วันพระ is Monday?

```python
bud_list[0].find('div', class_='bud-day-col').text
```

```python
countDay = [0]*7

for e in bud_list:
  dt = dateparser.parse(e.find('div', class_='bud-day-col').text)
  dt = dt.replace(year=dt.year-543)
  countDay[dt.weekday()] += 1

print(countDay[0])
```

**ANSWER** 21 days

```python
countDay
```

## Which day of the week that has the minimum number of วันพระ?

```python
print(countDay.index(min(countDay)))
```

**ANSWER** Tuesday

## Which day of the week that has the maximum number of วันพระ?

```python
print(countDay.index(max(countDay)))
```

**ANSWER** Sunday
