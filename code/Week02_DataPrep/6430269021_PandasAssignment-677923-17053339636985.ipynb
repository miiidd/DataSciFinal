## Compare "Sports" and "Comady", how many days that there are more total daily views of VDO in "Sports" category than in "Comady" category?

```python
import json
with open('data/US_category_id.json') as fd:
  category = json.load(fd)
category
```

```python
l = []
for item in category['items'] :
  l.append((int(item['id']),item['snippet']['title']))
category_df = pd.DataFrame(l,columns=['id','category'])
category_df
```

```python
vdo_df_with_cat = vdo_df.merge(category_df,left_on = 'category_id',right_on = 'id')
vdo_df_with_cat
```

```python
comedy_df = vdo_df_with_cat[vdo_df_with_cat.category =='Comedy'].groupby('trending_date').views.sum().reset_index(name='views_comedy')
comedy_df
```

```python
sport_df = vdo_df_with_cat[vdo_df_with_cat.category == 'Sports'].groupby('trending_date').views.sum().reset_index(name='views_sport')
sport_df
```

```python
merge_df = comedy_df.merge(sport_df,on = 'trending_date', how='inner')
merge_df[merge_df.views_sport > merge_df.views_comedy].shape[0]
```
