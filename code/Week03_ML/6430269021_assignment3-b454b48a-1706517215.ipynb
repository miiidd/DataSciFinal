```python
# 1. Load "ModifiedEdibleMushroom.csv" data from the link below (note: this data set has been preliminarily prepared.).
df = pd.read_csv('mushroom2020_dataset.csv')
```

```python
# [From step 1] before doing the data prep., how many "na" are there in "gill-size" variables?
df['gill-size'].isnull().sum()
```

```python
# 2. Drop rows where the target (label) variable is missing.
df = df.dropna(subset=['label'])
```

```python
df.shape
```

```python
# 3. Drop the following variables:

# 'id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate',

# 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type'

df.drop(['id','gill-attachment', 'gill-spacing', 'gill-size','gill-color-rate','stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring-rate','stalk-color-below-ring-rate','veil-color-rate','veil-type'], axis=1, inplace=True)
```

```python
df.shape
```

```python
# 4. Examine the number of rows, the number of digits, and whether any are missing.
df.info()
```

```python
# 5. Fill missing values by adding the mean for numeric variables and the mode for nominal variables.
print(df.isnull().sum())
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        df[col].fillna(df[col].mean(), inplace=True)
print(df.isnull().sum())
```

```python
df.info()
```

```python
# 6. Convert the label variable e (edible) to 1 and p (poisonous) to 0 and check the quantity. class0: class1
mapping_dict = {
    "label": {
        "e": 1,
        "p": 0,
    }
}
df = df.replace(mapping_dict)
df['label'].value_counts()
```

```python
# 7. Convert the nominal variable to numeric using a dummy code with drop_first = True.
nominal_columns = ["cap-shape",'cap-surface','bruises','odor','stalk-shape','ring-number','ring-type','spore-print-color','population','habitat']
dummy_df = pd.get_dummies(df[nominal_columns], columns=["cap-shape",'cap-surface','bruises','odor','stalk-shape','ring-number','ring-type','spore-print-color','population','habitat'], drop_first=True)
df = pd.concat([df, dummy_df], axis=1)
df.drop(nominal_columns,axis=1,inplace=True)
```

```python
df.shape
```

```python
dummy_df
```

```python
# 8. Split train/test with 20% test, stratify, and seed = 2020.
from sklearn.model_selection import train_test_split
y = df.pop('label')
X = df
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,test_size=0.20, random_state=2020)
print(X_train.shape)
print(X_test.shape)
print(y.value_counts()/y.shape[0])
print(y_train.value_counts()/y_train.shape[0])
print(y_test.value_counts()/y_test.shape[0])
```

```python
# 9. Create a Random Forest with GridSearch on training data with 5 CV.

# ​'criterion':['gini','entropy']
# 'max_depth': [2,3,6]
# 'min_samples_leaf':[2,5,10]
# 'N_estimators':[100,200]
# 'random_state': 2020

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 3, 6],
    'min_samples_leaf': [2, 5, 10],
    'n_estimators': [100, 200],
    'random_state': [2020]
}

grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=param_grid,
    cv=5,
)
grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
predictions = grid_search.predict(X_test)
```

```python
# 10. Predict the testing data set with confusion_matrix and classification_report.
print(classification_report(y_test,predictions,digits=2))
```

```python
print(confusion_matrix(y_test,predictions,labels=[0,1]))
```
